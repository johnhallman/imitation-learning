```python
def load_json(dir_file):
    with open(dir_file, 'r') as f:
        dir_data = json.load(f)
    return dir_data
```

```python
base_pend = load_json("experiment_data/final_base_pendulum1.json")
dagg_pend = load_json("experiment_data/final_dagg_pendulum1.json")
meil_pend = load_json("experiment_data/final_meil_pendulum3.json")
```

```python
pend_df = pd.DataFrame()
reward_pend_df = pd.DataFrame()
```

```python
columns = ['epoch', 'alg', 'expert_samples', 'agent_samples']
columns_reward = ['epoch', 'alg', 'reward', 'RE-loss']
for c in columns:
    pend_df[c] = None
for c in columns_reward:
    reward_pend_df[c] = None
```

```python
for k, v in meil_pend.items():
    pend_df = pend_df.append({'epoch':k, 'alg':"REIL", 'expert_samples':v['exp_s'], 
                              'agent_samples':v['pol_s']}, ignore_index=True)
    for r in v['rewards']:
        reward_pend_df = reward_pend_df.append({'epoch':k, 'alg':"REIL", 'reward':r}, ignore_index=True)
```

```python
cum = [0, 0]
for k, v in dagg_pend.items():
    cum[0] += v['exp_s']
    cum[1] += v['pol_s']
    pend_df = pend_df.append({'epoch':k, 'alg':"DAgger", 'expert_samples':cum[0], 
                              'agent_samples':cum[1]}, ignore_index=True)
    for r in v['rewards']:
        reward_pend_df = reward_pend_df.append({'epoch':k, 'alg':"DAgger", 'reward':r}, ignore_index=True)
```

```python
for k, v in base_pend.items():
    pend_df = pend_df.append({'epoch':k, 'alg':"Baseline", 'expert_samples':v['exp_s'], 
                              'agent_samples':0}, ignore_index=True)
    for r_l in v['rewards']:
        r = np.mean(r_l)
        reward_pend_df = reward_pend_df.append({'epoch':k, 'alg':"Baseline", 'reward':r}, ignore_index=True)
```

```python
pend_df.epoch = pend_df.epoch.astype('int64')
pend_df.expert_samples = pend_df.expert_samples.astype('int64')
pend_df.agent_samples = pend_df.agent_samples.astype('int64')
reward_pend_df.epoch = reward_pend_df.epoch.astype('int64')
reward_pend_df.reward = reward_pend_df.reward.astype('float64')
```

```python
def sample_graph(title, dir_meil, dir_dagg, dir_base):
    base_data = load_json(dir_base)
    dagg_data = load_json(dir_dagg)
    meil_data = load_json(dir_meil)
    
    pend_df = pd.DataFrame()
    for c in ['epoch', 'alg', 'expert_samples', 'agent_samples']:
        pend_df[c] = None
    for k, v in meil_data.items():
        pend_df = pend_df.append({'epoch':k, 'alg':"REIL", 'expert_samples':v['exp_s'], 
                                  'agent_samples':v['pol_s']}, ignore_index=True)
    cum = [0, 0]
    for k, v in dagg_pend.items():
        cum[0] += v['exp_s']
        cum[1] += v['pol_s']
        pend_df = pend_df.append({'epoch':k, 'alg':"DAgger", 'expert_samples':cum[0], 
                                  'agent_samples':cum[1]}, ignore_index=True)
    for k, v in base_pend.items():
        pend_df = pend_df.append({'epoch':k, 'alg':"Baseline", 'expert_samples':v['exp_s'], 
                                  'agent_samples':0}, ignore_index=True)
            
    pend_df.epoch = pend_df.epoch.astype('int64')
    pend_df.expert_samples = pend_df.expert_samples.astype('int64')
    pend_df.agent_samples = pend_df.agent_samples.astype('int64')

    ax = sns.lineplot(x='epoch', y="expert_samples", hue='alg', data=pend_df).set_title(title) # hue='event'
    #ax = sns.lineplot(x='epoch', y="agent_samples", hue='alg', data=pend_df) # hue='event'
    return

def reward_graph(title, dir_meil, dir_dagg, dir_base):
    base_data = load_json(dir_base)
    dagg_data = load_json(dir_dagg)
    meil_data = load_json(dir_meil)
    
    reward_pend_df = pd.DataFrame()
    for c in ['epoch', 'alg', 'reward']:
        reward_pend_df[c] = None
    for k, v in meil_data.items():
        for r in v['rewards']:
            reward_pend_df = reward_pend_df.append({'epoch':k, 'alg':"REIL", 'reward':r}, ignore_index=True)
    for k, v in dagg_pend.items():
        for r in v['rewards']:
            reward_pend_df = reward_pend_df.append({'epoch':k, 'alg':"DAgger", 'reward':r}, ignore_index=True)
    for k, v in base_pend.items():
        for r_l in v['rewards']:
            r = np.mean(r_l)
            reward_pend_df = reward_pend_df.append({'epoch':k, 'alg':"Baseline", 'reward':r}, ignore_index=True)
    reward_pend_df.epoch = reward_pend_df.epoch.astype('int64')
    reward_pend_df.reward = reward_pend_df.reward.astype('float64')
    
    ax = sns.lineplot(x='epoch', y="reward", hue='alg', data=reward_pend_df).set_title(title) # hue='event'
    return
```

```python
reward_graph("Reward over time, env: HalfCheetah-v2",  "experiment_data/final_meil_cheetah1.json", 
             "experiment_data/final_dagg_cheetah1.json", "experiment_data/final_base_cheetah1.json")
```

```python
def KL_graph(title, dir_meil):
    meil_data = load_json(dir_meil)
    
    reward_pend_df = pd.DataFrame()
    for c in ['epoch', 'RE_loss']:
        reward_pend_df[c] = None
    for k, v in meil_data.items():
        for r in v['RE-loss']:
            reward_pend_df = reward_pend_df.append({'epoch':k, 'RE_loss':r}, ignore_index=True)
    reward_pend_df.epoch = reward_pend_df.epoch.astype('int64')
    reward_pend_df.RE_loss = reward_pend_df.RE_loss.astype('float64')
    
    ax = sns.lineplot(x='epoch', y="RE_loss", data=reward_pend_df).set_title(title) # hue='event'
    return
```

```python
KL_graph("Relative-entropy loss, env: HalfCheetah-v2", "experiment_data/final_meil_cheetah1.json")
```
